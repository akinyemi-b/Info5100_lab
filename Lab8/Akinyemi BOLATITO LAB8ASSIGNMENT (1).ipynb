```python
# Import the necessary libraries
import pandas as pd

# Load the dataset
file_path = 'DESKTOP/Red_wine_cleaned.csv'  # Adjust path as needed
data = pd.read_csv(file_path)

# Display the first few rows to understand its structure
print("First 5 rows of the dataset:")
print(data.head())

# Display a summary of the dataset
print("\nDataset Info:")
data.info()

# Show basic statistics for numerical columns
print("\nDescriptive Statistics:")
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Drop unnecessary columns (if any)
data = data.drop(columns=['Unnamed: 0'], errors='ignore')  # Ignores if 'Unnamed: 0' doesn't exist

# Display dataset info and a preview
print(data.info())
data.head()
```

```python
# Displaying the correlation matrix of all numerical columns
print("Correlation Matrix:")
print(data.corr())
```

```python
# Displaying the correlation matrix of all numerical columns
print("Correlation Matrix:")
print(data.corr())
```

```python
# Displaying the first few rows of 'alcohol' and 'quality'
print("Preview of 'alcohol' and 'quality':")
print(data[['alcohol', 'quality']].head())
```

```python
# Previewing the 'pH' and 'quality' columns
print("Preview of 'pH' and 'quality':")
print(data[['pH', 'quality']].head())
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the directory 'images' exists
if not os.path.exists('images'):
    os.makedirs('images')

# Plot the histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['quality'], bins=7, kde=False, color='skyblue')
plt.title('Distribution of Wine Quality', fontsize=16)
plt.xlabel('Quality', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y')
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/histogram_quality.png')  # File saved in the 'imagesOfChartsCreatedinipynbfiles' folder
# Save the chart as an image

plt.show()
```

```python
#Visualization 2: Correlation Heatmap
# Plot a heatmap to show correlations between features
plt.figure(figsize=(12, 10))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Wine Dataset', fontsize=16)
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/correlation_heatmap.png')
plt.show()

```

```python
#Visualization 3: Scatter Plot of Alcohol vs. Quality

# Scatter plot: Alcohol content vs. wine quality
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=data, color='purple')
plt.title('Alcohol vs. Wine Quality', fontsize=16)
plt.xlabel('Alcohol Content', fontsize=14)
plt.ylabel('Quality Rating', fontsize=14)
plt.grid(True)
plt.savefig('images/scatter_alcohol_quality.png')  # Save the chart as an image
plt.show()
```

```python
#Visualization 3: Scatter Plot of Alcohol vs. Quality

# Box plot: Distribution of pH values for different wine quality ratings
# Scatter plot of Alcohol vs. Quality
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=data, color='blue')
plt.title('Alcohol vs. Quality', fontsize=16)
plt.xlabel('Alcohol', fontsize=14)
plt.ylabel('Quality', fontsize=14)
plt.grid(True)

# Save the chart as an image
if not os.path.exists('images'):
    os.makedirs('images')
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/scatterplot_alcohol_quality.png')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='quality', y='pH', data=data)  # No palette used
plt.title('Boxplot of pH by Wine Quality', fontsize=16)
plt.xlabel('Quality', fontsize=14)
plt.ylabel('pH', fontsize=14)
plt.grid(axis='y')

# Save the chart
if not os.path.exists('images'):
    os.makedirs('images')
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/boxplot_quality_ph.png')
plt.show()

```

```python
#to see pairwise relationships between several features.
sns.pairplot(data[['quality', 'alcohol', 'pH', 'citric acid', 'fixed acidity']])
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/Pairplot_quality_ph.png')
plt.show()
```

```python
#to see how a feature varies across different categories (e.g., quality levels
sns.boxplot(x='quality', y='alcohol', data=data, hue='quality', palette='Set2', legend=False)
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/boxplotcoloured_quality_ph.png')
```

```python
#to visualize the distribution of categories, such as how often each quality rating occurs
sns.barplot(x='quality', y='alcohol', data=data)
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/barplot_quality_ph.png')
plt.show()
```

```python
# to compare multiple features across subsets of data (e.g., how 'alcohol' varies for each 'quality').
g = sns.FacetGrid(data, col='quality')
g.map(sns.histplot, 'alcohol')
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/Histoplot_quality_ph.png')
plt.show()
```

```python
#to visualize the relationship between two continuous variables
sns.lineplot(x='quality', y='alcohol', data=data)
plt.savefig('C:/Users/akiny/Desktop/Info5100-lab/Lab8/imagesOfChartsCreatedinipynbfiles/lineplot_quality_ph.png')
plt.show()
```
